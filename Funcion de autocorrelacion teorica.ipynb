```python
from statsmodels.tsa.arima_process import arma_acf

# Calcular la FAC TEORICA de  (1 + 0.5B - 0.2B^2)W_t = a_t

# Definir los coeficientes del polinomio autoregresivo y del polinomio de medias móviles
phi = [1, -0.1, -0.5]
theta = [1]

# Definir los coeficientes de la varianza blanca
sigma = 1

# Calcular la FAC teórica
acf = arma_acf(ar=phi, ma=theta, lags=10)
print(acf)

# Graficar la FAC teórica
import matplotlib.pyplot as plt
plt.stem(acf)
plt.title('ACF teórica')
plt.show()

# Calcular la FAC teorica de  (1 + 0.7B - 0.1B^2)a_t = W_t
# 

# Definir los coeficientes del polinomio autoregresivo y del polinomio de medias móviles
# 
phi = [1]
theta = [1, 0.7, -0.1]

# Definir los coeficientes de la varianza blanca
sigma = 1

# Calcular la FAC teórica
acf = arma_acf(ar=phi, ma=theta, lags=10)
print(acf)

# Graficar la FAC teórica
import matplotlib.pyplot as plt
plt.stem(acf)
plt.title('ACF teórica')
plt.show()

```
