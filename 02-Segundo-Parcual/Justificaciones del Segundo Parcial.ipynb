```python
# partial autocorrelation function

menosintervalo = -2*np.sqrt(1/N)
masintervalo = 2*np.sqrt(1/N)
```

```python
print(menosintervalo, masintervalo)
```

### **Respuesta**

Entregar el desarrollo escrito de las pruebas de significancia y la justificación del modelo y región admisible correspondiente.



```python
def significancia(N, d, rho1, rho2, rho3):
    return 2 * np.sqrt(1 / (N-d) * (1 + 2 * (rho1**2 + rho2**2 + rho3**2)))
```

```python
def intervalo(N, d):
    return 2*np.sqrt(1/ (N-d) )
```

```python
vr1 = significancia(N, d, rho1, rho2, rho3)

if abs(r1) > vr1:
    print("rho 1 es significativo")
    print(r1, vr1)
    rho1 = r1
```

```python
vr2 = significancia(N, d, rho1, rho2, rho3)

if abs(r2) > vr2:
    print("rho 2 es significativo")
    print(r2, vr2)
    rho2 = r2
else:
    print("rho 2 no es significativo")
    print(r2, vr2)
```

```python
vr3 = significancia(N, d, rho1, rho2, rho3)

if abs(r3) > vr3:
    print("rho 3 es significativo")
    print(r3, vr3)
    rho3 = r3
else:
    print("rho 3 no es significativo")
    print(r3, vr3)
```

### **Respuesta**

```python
def FACP(N, d, r1, r2, r3, rho1=0, rho2=0, rho3=0):

    vr1 = significancia(N, d, rho1, rho2, rho3)
    if abs(r1) > vr1:
        print("rho 1 es significativo")
        print(r1, ">", vr1)
        rho1 = r1
    else:
        print("rho 1 no es significativo")
        print(r1,  "<", vr1)

    vr2 = significancia(N, d, rho1, rho2, rho3)
    if abs(r2) > vr2:
        print("rho 2 es significativo")
        print(r2, ">", vr2)
        rho2 = r2
    else:
        print("rho 2 no es significativo")
        print(r2, "<",  vr2)

    vr3 = significancia(N, d, rho1, rho2, rho3)
    if abs(r3) > vr3:
        print("rho 3 es significativo")
        print(r3, ">",vr3)
        rho3 = r3
    else:
        print("rho 3 no es significativo")
        print(r3,"<",  vr3)
```

## **Cuarto Ejercicio**

Se simuló una serie de tiempo integrada de orden 0 con 1000 observaciones, a continuación se muestran los valores de las autocorrelaciones simples y parciales muestrales, realice las pruebas de significancia necesarias e identifique un modelo que corresponda adecuadamente a los resultados de dichas pruebas. Entregar el desarrollo escrito de las pruebas de significancia y la justificación del modelo y región admisible correspondiente.



```python
def significancia(N, d, rho1, rho2, rho3, rho4, rho5):
    return 2 * np.sqrt(1 / (N-d) * (1 + 2 * (rho1**2 + rho2**2 + rho3**2 + rho4**2 + rho5**2)))
```

```python
def FACP5(N, d, r1, r2, r3, r4, r5, rho1=0, rho2=0, rho3=0, rho4=0, rho5=0):

    vr1 = significancia(N, d, rho1, rho2, rho3, rho4, rho5)
    if abs(r1) > vr1:
        print("rho 1 es significativo")
        print(abs(r1), ">", vr1)
        rho1 = r1
    else:
        print("rho 1 no es significativo")
        print(abs(r1),  "<", vr1)

    vr2 = significancia(N, d, rho1, rho2, rho3, rho4, rho5)
    if abs(r2) > vr2:
        print("\nrho 2 es significativo")
        print(abs(r2), ">", vr2)
        rho2 = r2
    else:
        print("\nrho 2 no es significativo")
        print(abs(r2), "<",  vr2)

    vr3 = significancia(N, d, rho1, rho2, rho3, rho4, rho5)
    if abs(r3) > vr3:
        print("\nrho 3 es significativo")
        print(abs(r3), ">",vr3)
        rho3 = r3
    else:
        print("\nrho 3 no es significativo")
        print(abs(r3),"<",  vr3)

    vr4 = significancia(N, d, rho1, rho2, rho3, rho4, rho5)
    if abs(r4) > vr4:
        print("\nrho 4 es significativo")
        print(abs(r4), ">",vr4)
        rho4 = r4
    else:
        print("\nrho 4 no es significativo")
        print(abs(r4),"<",  vr4)

    vr5 = significancia(N, d, rho1, rho2, rho3, rho4, rho5)
    if abs(r5) > vr5:
        print("\nrho 5 es significativo")
        print(abs(r5), ">",vr5)
        rho5 = r5
    else:
        print("\nrho 5 no es significativo")
        print(abs(r5),"<",  vr5)
```

```python
FACP5(1000, 0, -0.209, 0.172, -0.007, 0.058, -0.016)
```

```python
intervalo(1000, 0)
```

### **Respuesta**

- No hay diferencias.
- Solo las primera dos autocorrelacion simples son significativa.
- Muchas autocorrelaciones parciales son significativas.

Entonces es un $\text{ARIMA(0, 0, 2)}$.
