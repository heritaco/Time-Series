```python
import numpy as np
import matplotlib.pyplot as plt
import mpld3

# Parameters
t = np.arange(1, 201)
mean = 0
variance = 0.25

# Generate Nt
Nt = np.random.normal(mean, np.sqrt(variance), len(t))

# Generate Xt
Xt = np.cos(t / 10) + Nt

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(t, Xt, label='Xt')
ax.plot(t, np.cos(t / 10), 'r', label='cos(t/10)')
ax.set_title('Simulated Values of the Series $X_t = \cos(t/10) + N_t$')
ax.set_xlabel('t')
ax.set_ylabel('$X_t$')
ax.legend()
ax.grid(True)

# Display the plot in HTML
mpld3.display()
```

```python
import numpy as np

import matplotlib.pyplot as plt

# Parameters
t = np.arange(1, 201)
mean = 0
variance = 0.25

# Generate Nt
Nt = np.random.normal(mean, np.sqrt(variance), len(t))

# Generate Xt
Xt = np.cos(t / 10) + Nt

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(t, Xt, label='Xt')
plt.plot(t, np.cos(t / 10), 'r', label='cos(t/10)')
plt.title('Simulated Values of the Series $X_t = \cos(t/10) + N_t$')
plt.xlabel('t')
plt.ylabel('$X_t$')
plt.legend()
plt.grid(True)
plt.show()
```
